```python
# Load libraries
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from glob import glob
import pandas as pd

%matplotlib inline
```

# Cut the mosaic into smaller images

```python
image = np.load('/data/weed_data/160721_RGB_mosaic.npy')
image.shape
```

```python
# Reduce the image a bit 
redImage = image[:,4200:32000,2200:29000]
```

```python
# Rotate the image ~ 7 degrees
shape= redImage.shape[1:]
center = (27800//2, 26800//2)
matrix = cv2.getRotationMatrix2D(center, -7, 1.0)
rotated_out = np.zeros(shape=(26800,27800,3), dtype='uint8')

for i in range(3): 
    input_band = redImage[i].copy()
    rotated_out[:,:,i] = cv2.warpAffine(input_band,
                                        matrix,
                                        shape,
                                        flags = cv2.INTER_NEAREST,
                                        borderMode= cv2.BORDER_CONSTANT)
    
plt.imshow(rotated_out[:,:,1])
```

```python
# Cut into a huge square
redImage2 = rotated_out[1400:25000,1400:24500,:]
print(redImage2.shape)
plt.imshow(redImage2[:,:,1])
```

```python
redImage2.shape[:2]
```

```python
# split them into equal sized of 500:500 pixels 
# copied from the chunk_maker nb
def crop_chunks(img, img_path, out_path, min_shape, dims):
    #dims is the desired chunk size - height and width
    basename = os.path.basename(img_path)
    count = 0
    
    # are the images of the same size?
    if min_shape == 0:
        shape = img.shape[:2]
    else:
        shape = min_shape[:2]
    
    Row = int(round(shape[0]/dims))
    Column = int(round(shape[1]/dims))
    print(Row, Column)

    for row in range(Row):        
        for column in range(Column):
            # cut the chunk from the np array
            chunk = img[row*dims: (row+1)*dims,
                        column*dims:(column+1)*dims,
                       :]
            # name the output file
            new_file = basename.replace('.npy', f"_{count}.npy")
            output_file = os.path.join(out_path, new_file)
            np.save(output_file, chunk)
            count +=1
```

```python
# Cut into 500x500 pieces
outpath = '/data/weed_data/mosaic_prediction/npy_images'
crop_chunks(redImage2, "160721_RGB_mosaic.npy", outpath,0, 500)
```

```python
# Save as JPEGs
from PIL import Image

# Transform np into JPEG and save
path = "/data/weed_data/mosaic_prediction/npy_images/"

image_files = glob(f"{path}*npy")
image_files.sort()
```

```python
for i in range(len(image_files)):
    array = np.load(image_files[i])
#     array = np.transpose(array,(1,2,0)) #change order to x,y,channels
    im = Image.fromarray(array)
    im = im.convert("RGB")
    new_folder = os.path.dirname(path).replace("npy", "jpeg")
    new_filename = os.path.basename(image_files[i]).replace("npy", "jpeg")
    if not os.path.exists(new_folder):
        os.mkdir(new_folder)
    im.save(f"{new_folder}/{new_filename}") 
```
