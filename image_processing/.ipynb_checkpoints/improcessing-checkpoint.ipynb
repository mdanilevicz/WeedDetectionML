# Process field-1 and field-2 images


The field trial images were cut into plots using the "extract_plots.sh" bash script.

The functions described here are used for:
* Converting images from GeoTiff to numpy arrays.
* Rotating the images.
* Export as numpy array and JPEG.
* Segmentation of soil and plant pixels using a vegetation index.
* Use the bounding box labels for discriminating between crop and weed pixels.
* Exporting the images and segmentation masks.

# Rotate the plots


```python
def threshold_image(image):
    # Find the boundaries of the band
    thresh = cv2.adaptiveThreshold(image, 
                                   255, 
                                   cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY, 
                                   11, 2)
    # Get the coordenates for the bounding box
    coords = np.column_stack(np.where(thresh == 0)) # pixels = 0 mark the plot contour
    rect = cv2.minAreaRect(coords)
    angle = rect[-1] # find the rectangle with min area passing through the points array

    if angle < -45:
#         print(f" angle is {angle}")
        angle = -(90 + angle)
    else:
        angle = -angle

    # Rotation matrix for the image
    (h, w) = image.shape
    center = (w // 2, h // 2)
    # calculate the affine matrix of 2D rotation
    M = cv2.getRotationMatrix2D(center, angle, 1.0) 
    
    return angle, coords, M, rect
```

```python
def rotate_rgb_msi(img_path):
    # Get MSI file
    dataset = gdal.Open(img_path)
    msi_ds = dataset.ReadAsArray()
    
    # Create output folder
    name = dataset.GetDescription()
    basename = os.path.basename(name)
    basename =  os.path.splitext(basename)[0]
    fname = os.path.dirname(name)+"_rotated"
    if not os.path.exists(fname):
        os.mkdir(fname)

    # Get one band to find the threshold  
    band = msi_ds[1].copy() # copy one band from the 5 channel numpy array
    band = (band*255).astype('uint8') # adaptivethreshold only accepts 8bit 1 channel images
    angle, coords, M, rect = threshold_image(band)

    # Define the variables for cropping based on the rectangle
    center = tuple(map(int,(rect[0][1], rect[0][0])))
    size = tuple(map(int, (rect[1][1], rect[1][0]))) # size order = width, height
    
    # for band in range(n_bands):
    w_rect, h_rect = size[0], size[1]
    n_bands, h_msi, w_msi = msi_ds.shape
    
    # Create the array object in which the cropped bands will be saved
    crop_size = ( w_rect,(h_rect), n_bands)
    msi_out = np.zeros(shape=crop_size, dtype='uint8')

    # Process the MSI
    for band_index in range(n_bands):
        # Rotate the selected band
        input_band = msi_ds[band_index].copy()
        rotated_band = cv2.warpAffine(input_band, 
                                      M, ( w_msi, h_msi),
                                      flags=cv2.INTER_NEAREST,
                                      borderMode=cv2.BORDER_CONSTANT)
        rotated_band = (rotated_band).astype('uint8')
        
        # Crop the selected band
        cropped_band = cv2.getRectSubPix(rotated_band, (crop_size[1], crop_size[0]), center)  
        # Save the output to np array ds
        msi_out[:,:,band_index] = cropped_band
        
    # name the output file
    output_path_MSI = os.path.join(fname,basename)
    np.save(output_path_MSI, msi_out)
```

## Rotate and visualise

```python
# Make a list of all images to be processed in a directory
image_files = glob(f"{ft1}*tif*")                       
# Rotate all images in the list
for image in image_files:
    rotate_rgb_msi(image)
```

# Save as jpeg

```python
from PIL import Image
```

Because the images do not have the same size between collection dates, we will have a dynamic jpeg size.

```python
# Make it RGB
def create_rgb(img):
    img = np.load(img)
    RGB = np.zeros((img.shape[0], img.shape[1], 3))
    RGB[:,:,0] = RGB[:,:,2]
    RGB[:,:,1] = RGB[:,:,1]
    RGB[:,:,2] = RGB[:,:,0]
    
    RGB = RGB.astype("uint8")
    
    return RGB
```

## Segmentation of Plant - Soil

```python
# Calculate the VI for masks - In this case it is CIVE
def create_soilplant_maks(im):
    CIVE = 0.441*im[:,:,0] - 0.881*im[:,:,1] + 0.385*im[:,:,2] +18.78745
    otsu_threshold, image_result = cv2.threshold(CIVE.astype('uint8'), 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return otsu_threshold, image_result


otsu_thresh = []
for i in ft1_1:
    im = cv2.imread(i)
    im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
    otsu_threshold, image_result = create_soilplant_maks(im)
    path = os.path.dirname(i).replace('manuallabel_makesense', 'inputImages_Masks')
    filename = os.path.basename(i).replace('.jpeg',"")
    if not os.path.exists(path):
        os.makedirs(path)
    # save the image as npy
    np.save(f"{path}/{filename}_image.npy", im)
    # save the soil mask as npy
    np.save(f'{path}/{filename}_soil_mask.npy', image_result)
    otsu_thresh.append(otsu_threshold)
    
a = pd.Series(otsu_thresh)
a.to_csv(f'{path}/otsu_threshold.csv')
```

## Segmentation Crop - Weed

```python
# Get the bounding box labels
with open('/data/weed_data/manuallabel_makesense/F1_field_/labels_f1_fieldtrial_complete.json') as f:
    ft1_labels =  json.load(f)
ft1_images = sorted(glob("/data/weed_data/inputImages_Masks/F1_field/*_image.npy"))
ft1_masks = sorted(glob("/data/weed_data/inputImages_Masks/F1_field/*_soil_mask.npy"))
```

```python
import PIL.Image
import PIL.ImageDraw
import skimage.io

def polygons_to_mask(img_shape, polygons):
    mask = np.zeros(img_shape, dtype=np.uint8)
    mask = PIL.Image.fromarray(mask)
    xy = list(map(tuple, polygons))
    PIL.ImageDraw.Draw(mask).polygon(xy=xy, outline=1, fill=1)
    mask = np.array(mask, dtype=bool)
    return mask

def get_segmentation_coord(label_file, image_id=1):
    points=[]
    labels=[]
    half_mask=[]
    final_mask=[]
    
    anno = label_file['annotations']
    
    for ann in anno:
        im_id = ann['image_id']
        if im_id == image_id:
            seg1 = ann['segmentation']
            seg = seg1[0]
        
            for j in range(0,len(seg),2):
                x = seg[j]
                y = seg[j+1]
                mas = [x,y]
                half_mask.append(mas)
            final_mask.append(half_mask)
            half_mask=[]
        else:
            pass
    return final_mask

def weed_plant_labelling(soil_mask, weed_mask):
    final_mask = soil_mask.copy()
    final_mask[weed_mask] = np.where(final_mask[weed_mask] == 1, 2, 0)
    return final_mask

def remove_class(img, mask, obj_class=0):
    removed = np.zeros(img.shape, dtype='uint8')

    for i in range(img.shape[2]):
        band = img[:,:,i]
        band = np.where(mask == obj_class, 0, band)
        removed[:,:, i] = band
    return removed
```

```python
# Information needed to draw the weed mask
for i, im_label in enumerate(ft1_labels["images"]):
    filename = im_label["file_name"].replace(".jpeg", "")
    im_shape = [im_label["height"], im_label["width"]]
    
    # load the image and soil mask
    if os.path.exists(f'/data/weed_data/inputImages_Masks/F1_field/{filename}_image.npy'):
        rgb_im = np.load(f'/data/weed_data/inputImages_Masks/F1_field/{filename}_image.npy')
        soil_mask = np.load(f"/data/weed_data/inputImages_Masks/F1_field/{filename}_soil_mask.npy")
    
        # get coordenates
        im_id = im_label["id"]
        segcoord = get_segmentation_coord(ft1_labels, image_id=im_id)
        mask1 = polygons_to_mask(im_shape, segcoord[0])
        for j in range(1, len(segcoord)):
            maskany = polygons_to_mask(im_shape,segcoord[j])
            mask1 += maskany
        
        final_mask1 = weed_plant_labelling(soil_mask, mask1)
        np.save(f'/data/weed_data/inputImages_Masks/F1_field_/{filename}_weed_soil_crop_mask.npy', final_mask1)
    else:
        pass
```
